```python
import math
def loan_emi(amo,t,rate,down_pay=0):
    pay = amo-down_pay
    try:
        emi = pay *rate*((1+rate)**t)/(((1+rate)**t)-1)
    except ZeroDivisionError:
        emi= pay/t
    emi=math.ceil(emi)
    return emi
```

```python
loan7 = loan_emi(amo=1e5,
                t = 10*12,
                rate = 0
                )
print(loan7)
```

```python
loan5 = loan_emi(amo=1e5,
                t = 10*12,
                rate = 0.09/12)
print(loan5)
```
